```python
from flint import *
import numpy as np
import matplotlib.pyplot as plt
import time
plt.rcParams["figure.figsize"] = (20,20)
```

```python
def R(s,coefs):
    value = acb(1)
    for i in range(coefs.nrows()):
        value += coefs[i,0]*acb(i+2).pow(-s)
    return value
```

```python
def runTask(n,k):
    zeros = [0 for i in range(2*k)]
    ctx.dps = n
    print("Zeros number: ", k, " precision: ", n)
    with open("NImZetaZero40000_1_40000.val") as f:
        j = k-1
        for i in range(k):
            s = f.readline()[:n]
            zeros[i+k] = acb("0.5",s)
            zeros[j] = acb("0.5","-"+s)
            j = j - 1
    print(zeros[1])
    interval = np.arange(-6,0,0.01)
    res = [0 for item in interval]
    
    start_time = time.time()

    A = acb_mat(len(zeros), len(zeros))
    for i in range(len(zeros)):
        for j in range(len(zeros)):
            A[i,j] = acb(j+2).pow(-zeros[i])

    B = acb_mat([[-1] for i in range(len(zeros))])
    X = A.solve(B)
    for i in range(len(interval)):
        res[i] = R(interval[i],X).real
    print("--- %s seconds ---" % (time.time() - start_time))
    plt.axhline(y=0, color='green', linestyle='-')
    plt.plot(interval, res)
```

```python
runTask(1000,100)
```
